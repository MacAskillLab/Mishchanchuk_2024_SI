### Figure 1B

```python
trials_back, R, N, C = np.load('regression.npy', allow_pickle = True)
```

```python
plt.figure(figsize=(2,2))


sns.lineplot(x = trials_back, y = R,color = '0.2', err_style =  'bars', ci = 68)
sns.lineplot(x = trials_back, y = N, err_style =  'bars', ci = 68, color = '0.7')
sns.lineplot(x = trials_back, y = C, err_style =  'bars', ci = 68, color = '0.5')
plt.axhline(y=0, c = 'k', linestyle = '--',)

plt.xlabel('Trials back')

plt.xlim(2.5,0.95)
plt.ylim(-0.5,1.5)
plt.xticks([5,4,3,2,1], [-5,-4,-3,-2,-1])

plt.yticks([-0.5,0,0.5,1,1.5], [-0.5,0,0.5,1,1.5])
plt.ylabel('$\\beta$ coefficients')
sns.despine()
plt.tight_layout()
```

### Figure 1C

```python
trials, correct = np.load('switch_correct.npy', allow_pickle = True)
```

```python
sns.set_palette([pink])

ax = plt.figure(figsize=(2,2))
ax = sns.lineplot(x = trials, y = correct, ci = 68,linewidth=1, color = '0.4',label = 'data')


plt.axvline(x = -1, linestyle = '--', color = 'k')
plt.ylabel('p correct')
plt.xlabel('Trials from switch')
plt.xticks(np.arange(-7, 9,2), np.arange(-7,9,2))
plt.ylim(0,1)
plt.xlim(-7,7)
plt.yticks(np.arange(0, 1.05,0.2), np.arange(0, 1.1,0.2).round(1))

ax.get_legend().remove()

sns.despine()
plt.tight_layout()
```

### Figure 1G

```python
BIC_data = np.load('BIC.npy', allow_pickle = True)
```

```python
plt.figure(figsize = (1.5,2))

ax = sns.barplot(data = BIC_data, ci = 68, palette = [turq, pink], estimator = np.median)
sns.stripplot(data = BIC_data,
              color='0.3', jitter = False, size = 4)

plt.plot(BIC_data.T, color = '0.2', alpha = 0.3)

plt.ylim(0,70)

plt.xticks(rotation=0)
plt.xticks([0,1],['Q', 'SI'])
plt.xlabel('model')

plt.yticks(rotation = 0)
plt.ylabel('$\Delta$BIC')




sns.despine()
plt.tight_layout()
```

### Figure 1H
