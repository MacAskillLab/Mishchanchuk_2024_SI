### Figure 3B

```python
sham_mouse, lesion_mouse, sham_mouse_sess, lesion_mouse_sess = np.load('lesion_p_correct.npy', allow_pickle = True)
```

```python
plt.figure(figsize = (1.5,2))

sns.barplot(data = [sham_mouse, lesion_mouse], palette = ['0.4', 'w'], linewidth=1, edgecolor= ['0.4', '0.6'], estimator = np.mean)
sns.stripplot(data = [sham_mouse_sess, lesion_mouse_sess], color = '0.4', size = 3, alpha = 0.3)
sns.stripplot(data = [sham_mouse, lesion_mouse], color = '0.2', size = 3)


plt.yticks(rotation = 0)
plt.xticks([0,1],['-', '+'])
plt.xlabel('Lesion')
plt.ylabel('p correct')

plt.ylim(0.5, 0.8)

sns.despine()
plt.tight_layout()
```

```python
cpd_summary = np.load('cpd_summary.npy', allow_pickle=True)
```
